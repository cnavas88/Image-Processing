```python
import math
import matplotlib.pyplot as plt
from PIL import Image, ImageDraw
from skimage import color, data
import numpy as np
from scipy import ndimage
from scipy.misc import imresize
from pylab import ion, ioff
from scipy import misc

#escalo la imatge
scale_factor = 5
RGB=np.array(data.imread('casa.jpg'))
casa = imresize(RGB, ( RGB.shape[0]/scale_factor,RGB.shape[1]/scale_factor,3),interp='bilinear')
casaLAB = color.rgb2lab(casa.astype('float'))
casaColor = np.zeros(casa.shape)
for i in xrange(len(casaColor)):
    for j in xrange(len(casa[0])):
        casaColor[i][j] = casa[i][j]
gris  = np.sum(casa, axis=2) / 3

#fixo l'amplada i l'alcada
amplada = 10
alcada = 10
```

```python
#Funcions donades per Simone a la practica 3
class get_mouse_click():
    
    #Mouse interaction interface for radial distortion removal.
    def __init__(self, img):
      height, width = img.shape[:2]
      self.figure = plt.imshow(img, extent=(0, width, height, 0))
      plt.gray()
      plt.title('select the object to remove')
      plt.xlabel('Select sets of  points with left mouse button,\n'
                 'click right button to close the polygon.')
      plt.connect('button_press_event', self.button_press)
      plt.connect('motion_notify_event', self.mouse_move)

      self.img = np.atleast_3d(img)
      self.points = []
      self.centre = np.array([(width - 1)/2., (height - 1)/2.])

      self.height = height
      self.width = width

      self.make_cursorline()
      self.figure.axes.set_autoscale_on(False)

      plt.show()
      plt.close()

    def make_cursorline(self):
        self.cursorline, = plt.plot([0],[0],'r:+', linewidth=2,markersize=15,markeredgecolor='b')

    def button_press(self,event):
        # Register mouse clicks.
        if (event.button == 1 and event.xdata and event.ydata):
            self.points.append((event.xdata,event.ydata))
            plt.gca().lines.append(self.cursorline)
            plt.gca()
            self.make_cursorline()
        if (event.button != 1):
            self.points.append((self.points[0][0],self.points[0][1]))
            plt.close()
            return self.points

    def mouse_move(self,event):
        # Handle cursor drawing.
        pts_last_set=len(self.points)
        pts = np.zeros((pts_last_set+1,2))
        if pts_last_set > 0:
            # Line follows up to 3 clicked points:
            pts[:pts_last_set] = self.points[-pts_last_set:]
            # The last point of the line follows the mouse cursor
        pts[pts_last_set:] = [event.xdata,event.ydata]
        self.cursorline.set_data(pts[:,0], pts[:,1])
        plt.draw()
def compute_mask(width,height,polygon):
    
    img = Image.new('L', (width, height), 0)
    ImageDraw.Draw(img).polygon(polygon, outline=1, fill=1)
    mask = np.array(img)
    #mask100 = 100*np.ones([mask.shape[0],mask.shape[1]])
    #mask=mask*101
    #mask = 1. - mask; # switch 0s and 1s
    return mask
#Posa els forats a la imatge original
def ficarForats(mascara,imatgeOriginal):
    img = np.zeros(imatgeOriginal.shape)
    for i in xrange(len(mascara)):
        for j in xrange(len(mascara[0])):
            #comprovo si hi ha un forat a la mascara i si n'hi ha
            #poso el forat a la imatge principal
            if mascara[i][j] == 1:
                img[i][j] = -100
            else:
                img[i][j] = gris[i][j]
    return img
#Calcula la puntuacio
#Quan mes s'apropa al 0, pitjor es (1/diferencia)
#Quan es 1, es el valor exacte del forat.
def puntuacio(fila,col,mascara,casaForats,amplada,alcada):
    temp = 1
    amalca = 1.0/(amplada*alcada) #amplada i alcada, per normalitzar
    for i in xrange(alcada):
        for j in xrange(amplada):
            if casaForats[i+fila][j+col] == -100:
                return None
            else:
                temp += math.pow(mascara[i][j]-casaForats[i+fila][j+col],2)*amalca       
    return 1.0/temp
#genera la mascara de cada forat donada una amplada i alcada prefixada
def generarMascara(m,fila,col,alcada,amplada):
    k = list()
    for i in xrange(alcada):
        k.append(list())
        for j in xrange(amplada):
            if m[fila+i][j+col] == -100:
                k[i].append(0)
            else:
                k[i].append(m[fila+i][j+col])
    return k

def millorPuntuacio(gris,fila,col,mascara,casaForats,amplada,alcada):
    valor = 0
    dif = 0
    #recorro els forats
    for i in xrange(len(casaForats)):
        if len(casaForats)-i > alcada:
            for j in xrange(len(casaForats[0])):
                #comprovo si l'espai de la mascara es suficient al de l'element a comprovar
                if len(casaForats[0])-j > amplada:
                    puntuacioTemp = puntuacio(i,j,mascara,casaForats,amplada,alcada)
                    #si es pot calcular la puntuacio (None = forat)
                    if puntuacioTemp != None:
                        if puntuacioTemp > dif:
                            #directament, es la millor puntuacio possible.
                            #Es el valor que s'ha de posar al forat, doncs no hi ha diferencia
                            if puntuacioTemp == 0:
                                return i,j,gris[i][j]
                            else:
                                dif = puntuacioTemp
                                #de moment, es el millor candidat per copiar el pixel al forat
                                valor = i,j,gris[i][j]
                        
    return valor
```

```python
rdi = get_mouse_click(casaLAB[:,:,0])
mascara = compute_mask(casaLAB.shape[1],casaLAB.shape[0],rdi.points)
casaForats= ficarForats(mascara,gris)
cont = 0
#Recorro tots els forats.
for i in xrange(len(casaForats)):
    if len(casaForats)-i > alcada:
        for j in xrange(len(casaForats[0])):
            if len(casaForats[0])-j > amplada:
                #comprovo si es un forat
                if casaForats[i][j] == -100:
                    #posF,posC,casaForats[i][j] = millorPuntuacio(gris,i,j,generarMascara(gris,i,j,alcada,amplada),casaForats,amplada,alcada)
                    posF,posC,casaForats[i][j] = millorPuntuacio(gris,i,j,generarMascara(gris,i,j,alcada,amplada),casaForats,amplada,alcada)
                    casaColor[i][j] = casa[posF][posC]
                    #deso la imatge per generar un video
                    misc.imsave('imatges/'+str(cont+1)+'.png', casaForats.astype("uint8"))
                    cont  += 1
plt.subplot(121)
plt.imshow(gris.astype("uint8"),cmap="gray")
plt.title("Abans", size=20)

plt.subplot(122)
plt.imshow(casaForats.astype("uint8"),cmap="gray")
plt.title("Despres ", size=20)

plt.show()
```
